```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
# set url for scraping
url1 = 'https://fivethirtyeight.com/features/the-u-s-still-doesnt-know-how-to-track-a-pandemic/'
url2 = 'https://www.wired.com/story/how-to-stop-robots-becoming-racist/'
```

```python
# fetch content
html_five = requests.get(url1)
html_wired = requests.get(url2)
```

```python
# check status
html_five.status_code
html_wired.status_code
```

```python
soup_five = BeautifulSoup(html_five.text, 'html.parser')
```

```python
soup_wired = BeautifulSoup(html_wired.text, 'html.parser')
```
